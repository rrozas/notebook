```python
import glob
import pandas as pd
import ipywidgets as widgets
from ggplot import *
import seaborn as sbn
import matplotlib.pyplot as plt
```

```python
file = glob.glob("c:/Users/ROZASR/Documents/data/Ventes/Nov_2013-Mai_2015/x*")
with open("c:/Users/ROZASR/Documents/data/Ventes/Nov_2013-Mai_2015/En tête Extract 1.txt") as f : 
    header =[t.strip() for t in f.read().replace('"','').split(',')]
```

```python
file
```

```python
df = pd.DataFrame(columns=header)
for f in file:
    df = df.append(pd.read_csv(f, sep = "|", names=header))
```

```python
df
```

```python
df['Jour'] = pd.to_datetime(df.Jour,format="%Y%m%d")
```

```python
df["weekday"] = df.Jour.apply( lambda x : x.weekday())
```

```python
pd.datetime.weekday()
```

```python
df.groupby('weekday')["Quantité vendue permanent"].sum()
```

```python

rayonW = widgets.Select(options= df["Code rayon"].unique().tolist())

def f( rayon = 14 ):
    tmp = df[df["Code rayon"] == rayon] .groupby(["Code rayon", "Jour","weekday"])["Quantité vendue permanent"].sum().reset_index()
    g = ggplot(aes(x='Jour', y='Quantité vendue permanent', color = "weekday"), data=tmp)  + geom_point() \
    + scale_color_brewer(type='diverging', palette=7)
    display(g )
    
i = widgets.interact(f,rayon = rayonW )
```
